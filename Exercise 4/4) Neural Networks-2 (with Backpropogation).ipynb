```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import loadmat
from scipy import optimize
import utils

data=loadmat('ex4data1.mat')
X,y=data["X"], data['y'].ravel()

# set the zero digit to 0, rather than its mapped 10 in this dataset
# This is an artifact due to the fact that this dataset was used in 
# MATLAB where there is no index 0
y[y == 10] = 0

m=y.size #No. of training examples
#y.resize(m,1)
print(m, y.shape, X.shape)
```

```python
input_layer_size= 400 #s1
hidden_layer_size= 25 #s2
num_labels= 10 #s3

weights= loadmat("ex4weights.mat")
Theta1= weights['Theta1'] #dim(Theta1)= 25 x 401
Theta2= weights['Theta2'] #dim(Theta2)= 10 x 26

# swap first and last columns of Theta2, due to legacy from MATLAB indexing, 
# since the weight file ex3weights.mat was saved based on MATLAB indexing
Theta2= np.roll(Theta2,1, axis=0)
#UNROLLING PARAMETERS OF NN INTO THE Theta Variable so as to pass them to the optimising functions
Theta= np.concatenate([Theta1.ravel(), Theta2.ravel()], axis=0)
```

```python
def sigmoid(z):
    return 1/(1+ np.exp(-z))
```

```python
def nnCostFunction(Theta, input_layer_size, hidden_layer_size, num_labels, X, y, lambda_=0.0):    
    #lambda_ parameter is optional and if not put the default value used would be 0
    Theta1= Theta[0:(input_layer_size+1)*hidden_layer_size].reshape(hidden_layer_size, input_layer_size+1)
    Theta2= Theta[(input_layer_size+1)*hidden_layer_size:].reshape(num_labels, hidden_layer_size+1)
    
    #Part1 has feedforward part and returning the cost function. BEGIN:
    m=y.shape[0]
    if X.ndim==1:
        X=X[:, np.newaxis]
        
    X1=np.insert(X.copy(),0,1, axis=1)
    a2= sigmoid(np.matmul(Theta1, X1.transpose())) # dim(X)=m x input_layer_size+1 (due to bias unit)
    a2= np.insert(a2,0,1, axis=0) #dim(a2)= hidden_layer_size x m
    a3= sigmoid(np.matmul(Theta2,a2)) # dim(a3)= num_labels x m   a3=h
    
    y_new=np.zeros((num_labels,m))
    
    for i in range(m):
        y_new[y[i],i]=1
    
    J=-np.sum(y_new*np.log(a3)+ (1-y_new)*np.log(1-a3))/m + lambda_/(2*m)*(np.sum(Theta1[:,1:]**2) +np.sum(Theta2[:,1:]**2))
        
    #BACKPROPOGATION ALGORITHM STARTS BEGINS. Used Vectroised approach instead of the loop one for easy implementation
    delta3= a3-y_new #Calculating Error of output layer
    z2= np.matmul(Theta1, X1.transpose()) #dim(z2)= hidden_layer_size x m
    g_prime= sigmoid(z2)*(1-sigmoid(z2))
    
      #We don't use bias weights for BACKRPOP
    delta2= np.matmul(Theta2[:,1:].transpose(), delta3)*g_prime #dim(delta2)= hidden_layer_size x m
    
    #Now both Delta1 & Delta2 include biased a2 and X for calculations
    Delta2= np.matmul(delta3, a2.transpose()) # dim(Delta2)= s3 x s2 +1
    Delta1= np.matmul(delta2, X1) #dim(Delta1)= s2 x n+1  
    
    #Since Theta is local variable we set the bias weights to 0 for finally getting gradients after factoring in regularisation
    Theta_1, Theta_2= Theta1.copy(), Theta2.copy()
    
    Theta_1[:,0]=Theta_2[:,0]=0
                 
    D2= (Delta2 + lambda_*Theta_2)/m
    D1= (Delta1 + lambda_*Theta_1)/m
                 
    grad= np.concatenate([D1.ravel(), D2.ravel()], axis=0)
    
    return J, grad
```

```python
lambda_ = 0
J, _ = nnCostFunction(Theta, input_layer_size, hidden_layer_size,
                   num_labels, X, y, lambda_)
print('Cost at parameters (loaded from ex4weights)',J)
print('The cost should be about                   : 0.287629.')
```

```python
def randInitializeWeights(L_in, L_out, epsilon_init=0.12):
    #l_in= no. of units in the input layer, excluding BIAS
    #L_out no. of units in the ouput layer
    #To randomly intialise theta between +epsilon and -epsilon
    
    epsilon_init= np.sqrt(6)/np.sqrt(L_in+L_out)
    
    return (np.random.rand(L_out, L_in+1)*epsilon_init - epsilon_init) #increased L_in by 1 to account for bias weight as well
```

```python
initial_Theta1= randInitializeWeights(input_layer_size, hidden_layer_size)
initial_Theta2= randInitializeWeights(hidden_layer_size, num_labels)
initial_Theta= np.concatenate([initial_Theta1.ravel(), initial_Theta2.ravel()], axis=0)
```

## Learning Parameters using scipy.optimize

```python
costFunction= lambda p: nnCostFunction(p, input_layer_size, hidden_layer_size, num_labels,X, y.ravel(), lambda_= 1)

result= optimize.minimize(costFunction, initial_Theta, jac=True, method= 'TNC', options={'maxiter':500})

Theta= result.x

Theta1 = np.reshape(Theta[:hidden_layer_size * (input_layer_size + 1)],
                    (hidden_layer_size, (input_layer_size + 1)))

Theta2 = np.reshape(Theta[(hidden_layer_size * (input_layer_size + 1)):],
                    (num_labels, (hidden_layer_size + 1)))
```

```python
pred = utils.predict(Theta1, Theta2, X)
print('Training Set Accuracy',np.mean(pred == y) * 100,"%")
```
